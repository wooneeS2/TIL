```python
## 함수로 파생변수 만들기
## 조건문을 이용해 연비가 기준값을 넘으면 합격 or 불합격 하는 변수 만들기
```

```python
import numpy as np
## np.where()은 조건문을 한줄로 넣을 수 있는 함수
## 20이상이면 pass, 아니면 fail

mpg['test'] = np.where(mpg['total']>=20, 'pass','fail')
mpg.head()
```

```python
mpg['test'].value_counts()
```

```python
count_test = mpg['test'].value_counts()
count_test.plot.bar()
```

```python
count_test.plot.bar(rot=0) # x축 label을 수평으로 만들기
```

```python
#total 기준으로 A,B,C 등급 부여하기

mpg['grade'] = np.where(mpg['total']>=30,'A', np.where(mpg['total']>=20,'B','C'))

mpg.head()
```

```python
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
#알파벳순정렬
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```

```python
# ABCD 등급 만들기
# 범주의 수 -1 만큼 np.where() 중첩하기
#### 의문 : 꼭 이렇게 조건문을 쓸 수 밖에 없나,,? 보기도 복잡하고 중첩의 중첩인데 성능은 괜찮나?

mpg['grade2'] = np.where(mpg['total'] >=30, 'A',
                        np.where(mpg['total']>=25,'B',
                                np.where(mpg['total']>=20,'C','D')))
```

```python
mpg['size'] = np.where((mpg['category']=='compact')| 
                      ( mpg['category']== 'subcompact') |
                       (mpg['category']=='2seater'),
                        'small','large')
mpg['size'].value_counts()
```

```python
###분석도전
import pandas as pd

df_midwest = pd.read_csv('midwest.csv')
df_midwest.head()
```

```python
df_midwest.shape
```

```python
df_midwest.info()
```

```python
new_midwest = df_midwest.copy()
new_midwest= new_midwest.rename(columns={'poptotal':'total','popasian':'asian'})
new_midwest.head(5)
```

```python
#mpg['total'] = (mpg['cty']+mpg['hwy'])/2
new_midwest['asian_ratio'] = (new_midwest['asian'] / new_midwest['total'])*100
new_midwest.head(5)
```

```python
new_midwest['asian_ratio'].describe()
```

```python
new_midwest['asian_ratio'].plot.hist()
```

```python
import numpy as np
new_midwest['is_ave'] = np.where(new_midwest['asian_ratio']>=0.4872,'Large', 'Small')
new_midwest
```

```python
count_group = new_midwest['is_ave'].value_counts()
count_group
```

```python
count_group.plot.bar(rot=0)
```
